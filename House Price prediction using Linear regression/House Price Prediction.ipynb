```python
train.drop(['PoolQC','MiscFeature','Alley','Fence','Id'],axis=1,inplace=True)
train.shape
```

```python
train.isnull().sum().sort_values(ascending=False)
```

```python
X = train.drop(['SalePrice'], axis=1)
y = train[['SalePrice']].values.ravel() 
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

categorical_features = X.select_dtypes(include="object").columns   # take the categorical column names, we will use it for Transformer.
integer_features = X.select_dtypes(exclude="object").columns       # take the numerical column names.


for c in categorical_features:       # converting the labels into a numeric form for each categorical column
    lbl = LabelEncoder() 
    lbl.fit(list(X[c].values)) 
    X[c] = lbl.transform(list(X[c].values))

integer_transformer = Pipeline(steps = [
   ('imputer', SimpleImputer(strategy = 'mean')),   #fill the N/A values with the mean of each numerical column (SimpleImputer).
   ('scaler', StandardScaler())])                   #Standardize features by removing the mean and scaling to unit variance (StandardScaler).



categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent'))  #fill the N/A values with the most_frequent value of each categorical column (SimpleImputer).
])

preprocessor = ColumnTransformer(                         #final form of pipeline
   transformers=[
       ('ints', integer_transformer, integer_features),
       ('cat', categorical_transformer, categorical_features)])


X = preprocessor.fit_transform(X)                        #make the transform to the Features

X
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X,y, test_size=0.20, random_state=42,shuffle=False)

print('X_train:',X_train.shape,'y_train:',y_train.shape,'\nX_valid :',X_valid.shape,'y_valid:',y_valid.shape)
```
